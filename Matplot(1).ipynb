```python
fig,ax=plt.subplots()
x=np.random.randint(1,10,size=10)
y=10000*x
ax.plot(x,y)
ax.plot(x,x+3)
plt.plot(x,2*y)
plt.show()
```

```python
fig,ax=plt.subplots()
x=np.linspace(0,10,10000)
y=2*x
ax.plot(x,np.cos(x))
ax.plot(x,y,color=(1.0,0.2,0.3),linestyle='--')
ax.plot(x,np.sin(x),color='r',linestyle=':')
plt.show()
```

```python
plt.plot(x,np.cos(x),label='y=cos(x)')
plt.plot(x,np.sin(y),label='y=sin(x)')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.read_csv('hr.csv')
```

```python
df.head()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
x=df['GDP']
y=df['Score']
plt.scatter(x,y)
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP')
plt.ylabel('Happiness Score')
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(df['Year'],df['GDP'],df['Score'],s=30)
ax.set(xlabel='Year',ylabel='GDP',zlabel='Score')
plt.xticks(np.arange(2105,2020,step=1))
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(df['Year'],df['GDP'],df['Freedom'],s=30)
ax.set(xlabel='Year',ylabel='GDP',zlabel='Freedom')
plt.xticks(np.arange(2105,2020,step=1))
plt.show()
```

```python
import sklearn as sklearn
from sklearn.model_selection import train_test_split
from sklearn import datasets,metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df['GDP']
y=df['Score']
x=x.values.reshape(-1,1)
y=y.values.reshape(-1,1)
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.5,random_state=0)
```

```python
h=LinearRegression()
h.fit(X_train,Y_train)
y_pred=h.predict(X_test)
```

```python
print(y_pred.shape)
```

```python
fig,(ax1)=plt.subplots(1,figsize=(12,6))
ax1.scatter(X_test,Y_test,s=8)
plt.plot(X_test,y_pred,color='black',linewidth=2)
plt.show()
```
